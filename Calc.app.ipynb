```python
import PySimpleGUI as sg
from PySimpleGUI import *

sg.theme('Dark')

bt_ds_font=('century gothic',18)
bt_font=('century gothic',10)
bt_color=('black','#E5E4E2')
op_bt_color=('white','#045D5D')
bt_size=(5,2)

# menu bar
menu = [
        ['Calc.',['Standard','Scientific','Programmer','Currency','Length']],
        ['History'],
    	['Help']
       ]

# GUI Layout
layout = [
          [Menu(menu)],
          [Text('', size = (15, 2), font =bt_ds_font,background_color='#C0C0C0',text_color = 'black', key = 'input')],
          [ReadFormButton('C',button_color=op_bt_color,font=bt_font) ,ReadFormButton('CE',button_color=op_bt_color,font=bt_font),ReadFormButton('«',button_color=op_bt_color,font=bt_font),ReadFormButton('/',button_color=op_bt_color,font=bt_font)],
          [ReadFormButton('7',button_color=bt_color,font=bt_font), ReadFormButton('8',button_color=bt_color,font=bt_font), ReadFormButton('9',button_color=bt_color,font=bt_font), ReadFormButton('*',button_color=op_bt_color,font=bt_font)],
          [ReadFormButton('4',button_color=bt_color,font=bt_font), ReadFormButton('5',button_color=bt_color,font=bt_font), ReadFormButton('6',button_color=bt_color,font=bt_font), ReadFormButton('-',button_color=op_bt_color,font=bt_font)],
          [ReadFormButton('1',button_color=bt_color,font=bt_font), ReadFormButton('2',button_color=bt_color,font=bt_font), ReadFormButton('3',button_color=bt_color,font=bt_font), ReadFormButton('+',button_color=op_bt_color,font=bt_font)],
          [ReadFormButton('0',button_color=bt_color,font=bt_font), ReadFormButton('.',button_color=bt_color,font=bt_font), ReadFormButton('=',button_color=op_bt_color,font=bt_font,size=(12,2))],
          ]
  
# Set PySimpleGUI
form = FlexForm('Calculator', default_button_element_size = (5, 2),
                auto_size_buttons = False, grab_anywhere = False)
form.Layout(layout)
  
# Result Value
Result = ''
  
# Make Infinite Loop
while True:
    # Button Values
    button, value = form.Read()
      
    # Check Press Button Values
    if button == 'C':
        Result = ''
        form.FindElement('input').Update(Result)
    elif button=='«':
        Result = Result[:-1]
        form.FindElement('input').Update(Result)
    elif len(Result) == 16 :
        pass
      
   # Results
    elif button == '=':
        Answer = eval(Result)
        Answer = str(round(float(Answer),3))
        form.FindElement('input').Update(Answer)
        Result = Answer
          
    # close the window
    elif button == 'Quit'  or button == None:
        break
    else:
        Result += button
        form.FindElement('input').Update(Result)
```
